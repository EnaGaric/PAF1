```python
print("Unesi koordinate za dvije točke:")
print(" ")

def unos_koordinata():
    while True:
        try:
            x1 = float(input("x1: "))
            y1 = float(input("y1 :"))
            x2 = float(input("x2: "))
            y2 = float(input("y2 :"))
            return x1,y1,x2,y2
        except ValueError:
            print("Molim te da uneseš broj u decimalnom zapisu.\n")

x1, y1, x2, y2 = unos_koordinata()

def jednadzba_pravca(x1,y1,x2,y2):
    if x1 == x2:
        print("Pravac je vertikalan, x = {}".format(x1))
    else:
        k = (y2 - y1) / (x2 - x1)
        l = y1 - k*x1
        print("jednadzba pravca:")
        print("y = {}x + {}".format(k,l))

jednadzba_pravca(x1,y1,x2,y2)
```
